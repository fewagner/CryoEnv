```python
from cryoenv.buffers import ReplayBuffer
import numpy as np
```

```python
n_obs = 2
dim_hidd = 64
n_actions = 2
len_buffer = 1000

burn_in = 5
steps = 20
batch_size = 16

buffer = ReplayBuffer(len_buffer, 
                      (n_obs, ), 
                      n_actions)
```

```python
for n in [150]*15:
    next_state = np.zeros(n_obs)
    for i in range(n):
        state = next_state
        action = np.random.uniform(-1,1,size=n_actions)
        reward = np.random.normal()
        next_state = np.random.normal(-1,1,size=n_actions)
        terminal = np.random.uniform() < .01
        trajectory_idx = i
        buffer.store_transition(state, action, reward, next_state, terminal, trajectory_idx)
        if terminal:
            break
```

```python
for vals in zip(buffer.state_memory, buffer.next_state_memory, buffer.action_memory, buffer.reward_memory, buffer.terminal_memory, buffer.trajectory_idx):
    print('----------')
    for nameing, element in zip(['state', 'next state', 'action', 'reward', 'terminal', 'trajectory idx'], vals):
            print('{}: {}'.format(nameing, element))
```

```python
buffer.sample_trajectories(batch_size, steps, burn_in)
```
